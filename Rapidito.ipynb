##**Se Busca la Variable Principal**

```python
# prompt: analiza los datos para encontrar la variable principal

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the goal is to find the target variable for a potential prediction task
# or understand the relationships between variables.
# We can visually inspect the distributions and relationships.

# First, let's visualize the distribution of numerical features.
df.hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features', y=0.95)
plt.show()

# Next, let's look at the correlation matrix for numerical features.
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# For categorical features, we can visualize their value counts.
for col in df.select_dtypes(include='object').columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Value Counts for {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()

# Based on the context (e.g., predicting sales, customer behavior, etc.),
# the "principal variable" (main variable) would likely be the one you are trying to predict
# or the one that represents the primary outcome of interest.

# Without a specific task defined, it's difficult to definitively state the "principal variable".
# However, based on typical business scenarios for a pharmaceutical company dataset,
# potential principal variables could be:
# - 'Ventas': If the goal is to predict sales.
# - 'Numero de Pedidos': If the goal is to predict the volume of orders.
# - 'Rating' (if it existed): If the goal was customer satisfaction prediction.

# To determine the principal variable, you need to know the objective of the analysis.
# If this is exploratory, the "principal variables" might be those with interesting distributions,
# significant correlations with other variables, or the most unique categories.

print("\nAnalysis Summary:")
print("Review the histograms, correlation matrix, and count plots.")
print("Identify the variable that represents the main outcome or focus of your analysis.")
print("Common principal variables in this context might relate to sales or order volume if prediction is the goal.")
```
